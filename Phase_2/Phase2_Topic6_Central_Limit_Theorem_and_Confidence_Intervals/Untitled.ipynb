```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate values for the x-axis
x_values = np.linspace(-3, 3, 1000)

# Calculate the CDF values for the standard normal distribution
cdf_values = norm.cdf(x_values, loc=0, scale=1)

# Plot the CDF
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(x_values, cdf_values, label='CDF')
plt.title('Cumulative Distribution Function (CDF)')
plt.xlabel('Standard Deviations from the Mean')
plt.ylabel('Probability')
plt.legend()

# Calculate the PPF values for specific probabilities (e.g., 0.25, 0.50, 0.75)
ppf_values_25 = norm.ppf(0.25, loc=0, scale=1)
ppf_values_50 = norm.ppf(0.50, loc=0, scale=1)
ppf_values_75 = norm.ppf(0.75, loc=0, scale=1)

# Plot the PPF
plt.subplot(1, 2, 2)
plt.plot(x_values, cdf_values, label='CDF')
plt.scatter([ppf_values_25, ppf_values_50, ppf_values_75], [0.25, 0.50, 0.75], color='red', zorder=5, label='PPF Values')
plt.title('Percent-Point Function (PPF)')
plt.xlabel('Standard Deviations from the Mean')
plt.ylabel('Probability')
plt.legend()

plt.tight_layout()
plt.show()
```
